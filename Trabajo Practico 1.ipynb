```python
import pandas as pd
import statsmodels.api as sm
import numpy  as np
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("nb1-insurance.csv")
df.head()
```

```python
# Variable objetivo
Y = df["expenses"]
#Variables predictivas
X = df[["age","sex","bmi","children","smoker"]]
```

```python
pd.read
```

```python
columns_names = df.columns.values.tolist()
columns_names
```

```python
dummy_sex = pd.get_dummies(df["sex"], prefix="sex").iloc[:,1:]
dummy_sex.head()
```

```python
dummy_smoker = pd.get_dummies(df["smoker"], prefix="smoker").iloc[:,1:]
dummy_smoker.head()
```

```python
columns_names = df.columns.values.tolist()
df_new = df[columns_names].join(dummy_sex)
columns_names = df_new.columns.values.tolist()
df_new = df_new[columns_names].join(dummy_smoker)
df_new.head()
```

```python
feature_cols = ["age","bmi","children","sex_male","smoker_yes"]
# Variable objetivo
Y = df_new["expenses"]
#Variables predictivas
X = df_new[feature_cols]
```

```python
# Normalizando variables
scale = StandardScaler()
X_scaled = scale.fit_transform(X[feature_cols].as_matrix())
```

```python
# Estimacion variables sin normalizar
est = sm.OLS(Y, X).fit()
est.summary()
```

```python
# Estimacion variables normalizadas
est2 = sm.OLS(Y,X_scaled).fit()
est2.summary()
```

```python
print("Variables sin normalizar:\n",est.params)
print()
print("Varialbes normalizadas\n",est2.params)
```

```python
print("R^2 variable sin normalizar: ",est.rsquared)
print("R^2 varialbes normalizadas: ",est2.rsquared)
```

```python
def predict_expenses(age, bmi, children,sex_male,smoker_yes):
    return 198.399764*age  +  37.578382*bmi + 250.841441*children - 707.668406*sex_male + 23386.227638*smoker_yes
```

```python
def predict_expenses_normalize(age, bmi, children,sex_male,smoker_yes):
    return 3619.590806*age  +  1965.698187*bmi + 571.915447*children - 64.337115*sex_male + 9613.331140*smoker_yes
```

```python
df_new["prediction"] = predict_expenses(df_new["age"],df_new["bmi"],df_new["children"],df_new["sex_male"],df_new["smoker_yes"])
df_new["prediction_normalize"] = predict_expenses_normalize(df_new["age"],df_new["bmi"],df_new["children"],df_new["sex_male"],df_new["smoker_yes"])
```
